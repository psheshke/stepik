```python
n = int(input())

a = [[0 for j in range(n)] for i in range(n)]

i_min = 0
i_max = n-1
j_min = 0
j_max = n-1
k = 1
move_to = 0

i = 0
j = 0

while k <= n ** 2:
    if move_to == 0:
        while j != j_max + 1:
            a[i][j] = k
            k += 1
            j += 1
        i_min += 1
        i += 1
        j -= 1
        move_to += 1
    elif move_to == 1:
        while i != i_max + 1:
            a[i][j] = k
            k += 1
            i += 1
        j_max -= 1
        j -= 1
        i -= 1
        move_to += 1
    elif move_to == 2:
        while j != j_min - 1:
            a[i][j] = k
            k += 1
            j -= 1
        i_max -= 1
        i -= 1
        j += 1
        move_to += 1
    elif move_to == 3:
        while i != i_min - 1:
            a[i][j] = k
            k += 1
            i -= 1
        j_min += 1
        j += 1
        i += 1
        move_to = 0

for row in a:
    for column in row:
        print(column, end = ' ')
    print()
```
